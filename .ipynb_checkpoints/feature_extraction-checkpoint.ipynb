```python
def get_random_image(filepath):
  all_files = os.listdir(filepath)
  rand_image_file = random.choice(all_files)
  return os.path.join(filepath, rand_image_file)
```

```python
def hist_plot(rand_img, name):
  image = cv2.imread(rand_img)
  if len(image.shape) == 2:
      # Grayscale image
      channel_axis = None
  else:
      # Color image
      channel_axis = -1
  fd, hog_image = hog(image, orientations=9, pixels_per_cell=(8, 8),
                      cells_per_block=(2, 2), visualize=True, channel_axis=channel_axis)
  hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
  # Plot the original image
  plt.figure(figsize=(12, 6))
  plt.subplot(1, 3, 1)
  plt.title(f'Original Image: {name}')
  plt.imshow(image, cmap='gray')
  plt.axis('off')

  # Plot the HOG image
  plt.subplot(1, 3, 2)
  plt.title('HOG Image')
  plt.imshow(hog_image_rescaled, cmap='gray')
  plt.axis('off')

  # Compute and plot the histogram of intensities
  plt.subplot(1, 3, 3)
  plt.title('Histogram of Intensities')
  plt.hist(image.ravel(), bins=256, range=[0, 256])
  plt.xlabel('Pixel Intensity')
  plt.ylabel('Frequency')

  # Show all plots
  plt.tight_layout()
  plt.show()
```

```python
hist_plot(hairline_img, 'Hairline Fracture')
hist_plot(spiral_img, 'Spiral Fracture')
hist_plot(greenstick_img, 'Greenstick Fracture')
hist_plot(comminuted_img, 'Comminuted Fracture')
hist_plot(dislocation_img, 'Fracture Dislocation')
hist_plot(pathological_img, 'Pathological Fracture')
hist_plot(longitudinal_img, 'Longitudinal Fracture')
hist_plot(oblique_img, 'Oblique Fracture')
hist_plot(impacted_img, 'Impacted Fracture')
hist_plot(avulsion_img, 'Avulsion Fracture')
```

```python
import cv2
import os
import numpy as np
import pandas as pd
from skimage.feature import hog
from skimage import exposure
import tensorflow as tf

# Define paths for each fracture type
root = 'data/'
paths = {
    'Hairline Fracture': root + 'Hairline Fracture/Train',
    'Spiral Fracture': root + 'Spiral Fracture/Train',
    'Greenstick Fracture': root + 'Greenstick fracture/Train',
    'Comminuted Fracture': root + 'Comminuted fracture/Train',
    'Fracture Dislocation': root + 'Fracture Dislocation/Train',
    'Pathological Fracture': root + 'Pathological fracture/Train',
    'Longitudinal Fracture': root + 'Longitudinal fracture/Train',
    'Oblique Fracture': root + 'Oblique fracture/Train',
    'Impacted Fracture': root + 'Impacted fracture/Train',
    'Avulsion Fracture': root + 'Avulsion fracture/Train'
}

# Function to load and preprocess images
def load_image(image_path, image_size=(256, 256)):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, image_size)
    image = image / 255.0  # Normalize to [0,1] range
    return image

# Function to extract HOG features
def extract_hog_features(image, image_size=(256, 256)):
    image = image.numpy()  # Convert EagerTensor to NumPy array
    image = (image * 255).astype('uint8')
    fd = hog(image, orientations=9, pixels_per_cell=(8, 8),
             cells_per_block=(2, 2), visualize=False, channel_axis=-1)
    return fd

# Initialize lists to hold HOG features and labels
hog_features = []
labels = []

# Loop over the paths and extract HOG features
for label, (fracture_type, path) in enumerate(paths.items()):
    # List files in the directory
    files = tf.io.gfile.listdir(path)
    for file in files:
        image_path = os.path.join(path, file)
        # Load and preprocess the image
        image = load_image(image_path)
        # Extract HOG features
        fd = extract_hog_features(image)
        # Append features and label to the lists
        hog_features.append(fd)
        labels.append(label)

# Convert lists to NumPy arrays
hog_features = np.array(hog_features)
labels = np.array(labels)

# Optionally, save to a Pandas DataFrame for easier handling
df = pd.DataFrame(hog_features)
df['label'] = labels

# Save to a CSV file for later use
df.to_csv('hog_features_train.csv', index=False)

# Verify the shape of the feature matrix and labels
print('HOG Features shape:', hog_features.shape)
print('Labels shape:', labels.shape)
```
